```python
# Data Cleaning and making subsets of years
import pandas as pd
import matplotlib.pyplot as plt
from apriori import apriori

df = pd.read_csv("US_Accidents_Dec21_updated.csv")
df = df.drop('ID', axis=1)
df = df.drop('Zipcode', axis=1)
df = df.drop('Country', axis=1)
df = df.drop('Turning_Loop', axis=1)
df = df.drop('Civil_Twilight', axis=1)
df = df.drop('Nautical_Twilight', axis=1)
df = df.drop('Astronomical_Twilight', axis=1)
df = df.drop('Airport_Code', axis=1)
df = df.drop('Traffic_Calming', axis=1)
df = df.drop('Bump', axis=1)
df = df.drop('Railway', axis=1)
df = df.drop('Description', axis=1)
df = df.drop('No_Exit', axis=1)
df = df.drop('Roundabout', axis=1)
df = df.drop('End_Time', axis=1)
df = df.drop('Start_Lat', axis=1)
df = df.drop('Start_Lng', axis=1)
df = df.drop('End_Lat', axis=1)
df = df.drop('End_Lng', axis=1)
df = df.drop('Number', axis=1)



# Drop the year you want as a subset for Association Rule Mining
#rows = df[df['Start_Time'].str[:4] == '2016'].index
#df = df.drop(rows)

rows = df[df['Start_Time'].str[:4] == '2017'].index
df = df.drop(rows)
rows = df[df['Start_Time'].str[:4] == '2018'].index
df = df.drop(rows)
rows = df[df['Start_Time'].str[:4] == '2019'].index
df = df.drop(rows)
rows = df[df['Start_Time'].str[:4] == '2020'].index
df = df.drop(rows)
rows = df[df['Start_Time'].str[:4] == '2021'].index
df = df.drop(rows)

print(df.shape)
print(df.columns)


output_csv_file = "CSCI185ProjectDatasetY2016.csv"
df.to_csv(output_csv_file, index=False)
```

```python
# Association Rule Mining on the 2016 subset

import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori

#Read the file
df = pd.pandas.read_csv("CSCI185ProjectDatasetY2016.csv", header = None)

records = []
for i in range(0, len(df)):
    records.append([str(df.values[i,j]) for j in range(0, 27)])
association = apriori(records, min_support=0.1, min_confidence=0.1, min_lift=3)
association_results = list(association)
print(len(association_results))



support = []
confidence = []
# Extract support and confidence values from association rules
for item in association_results:
    support.append(item[1])
    confidence.append(item[2][0][2])

# Create a scatter plot of support vs. confidence
plt.figure(figsize=(10, 10))
plt.scatter(support, confidence, alpha=0.5)
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Association Rules: Support vs. Confidence - 2016")
plt.show()

#Class Code
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
